```python
import random
import numpy as np

import scipy.stats
from numpy.random import randn

import matplotlib.pyplot as plt

random.seed(1234)
np.random.seed(1234)

%matplotlib inline
```

```python
num_data = 200
e_z = randn(num_data)
e_y = randn(num_data)

Z = e_z
Y = 2 * Z + e_y
```

```python
np.corrcoef(Z, Y)
```

```python
Z_std = scipy.stats.zscore(Z)
Y_std = scipy.stats.zscore(Y)
```

```python
plt.scatter(Z_std, Y_std)
```

```python
# 擬似相関：因果が逆

num_data = 200
e_z = randn(num_data)
e_y = randn(num_data)

Y = e_y
Z = 2*Y + e_z


np.corrcoef(Z, Y)
```

```python
z_std = scipy.stats.zscore(Z)
y_std = scipy.stats.zscore(Y)

plt.scatter(z_std, y_std)
```

```python
#擬似相関: 共通の原因（交絡）

num_data = 200
e_x = randn(num_data)
e_y = randn(num_data)
e_z = randn(num_data)

Z = 3.3  * e_x + e_z
Y = 3.3 * e_x + e_y


np.corrcoef(Z, Y)
```

```python
z_std = scipy.stats.zscore(Z)
y_std = scipy.stats.zscore(Y)


np.corrcoef(Z, Y)
```

```python
#合流点での選抜

num_data = 600
e_x = randn(num_data)
e_y = randn(num_data)

x = e_x
y = e_y

plt.scatter(x, y)
```

```python
np.corrcoef(x, y)
```

```python
z = x + y

x_new = np.array([])
y_new = np.array([])
for i in range(num_data):
    if z[i] > 0.0 :
        x_new = np.append(x_new, x[i])
        y_new = np.append(y_new, z[i])
        
plt.scatter(x_new, y_new)
```

```python
np.corrcoef(x_new, y_new)
```
